2. Tranform Data

```python
#enrollies_data
enrollies_data.info()
```

```python
#set full_name as string
enrollies_data['full_name'] = enrollies_data['full_name'].astype('string')
#set city as string
enrollies_data['city'] = enrollies_data['city'].astype('string')
#set gender as category type
enrollies_data['gender'] = enrollies_data['gender'].astype('category')
#find mode gender
gender_mode = enrollies_data['gender'].mode()[0]
#Fill null by gender_mode
enrollies_data['gender'] = enrollies_data['gender'].fillna(gender_mode)
enrollies_data.info()
```

```python
#education_data
education_data.info()
education_data.sample(10)
```

```python
#tranform data
education_data['major_discipline'] = education_data['major_discipline'].astype('category')
education_data['education_level'] = education_data['education_level'].astype('category')
education_data['enrolled_university'] = education_data['enrolled_university'].astype('category')
```

```python
# Assuming 'education_data' is a pandas DataFrame you have loaded or created earlier
education_data['major_discipline'] = education_data['major_discipline'].fillna(education_data['major_discipline'].mode()[0])
education_data['education_level'] = education_data['education_level'].fillna(education_data['education_level'].mode()[0])
education_data['enrolled_university'] = education_data['enrolled_university'].fillna(education_data['enrolled_university'].mode()[0])
education_data.info()
```

```python
#work_experience
work_experience.info()
work_experience.sample(10)
```

```python
#tranform dataa work experience
work_experience['experience'] = work_experience['experience'].astype('category')
work_experience['relevent_experience'] = work_experience['relevent_experience'].astype('category')
work_experience['last_new_job'] = work_experience['last_new_job'].astype('category')
work_experience['company_size'] = work_experience['company_size'].astype('category')
work_experience['company_type'] = work_experience['company_type'].astype('category')
#Fill null dataa
work_experience['experience'] = work_experience['experience'].fillna(work_experience['experience'].mode()[0])
work_experience['company_size'] = work_experience['company_size'].fillna(work_experience['company_size'].mode()[0])
work_experience['company_type'] = work_experience['company_type'].fillna(work_experience['company_type'].mode()[0])
work_experience['last_new_job'] = work_experience['last_new_job'].fillna(work_experience['last_new_job'].mode()[0])
work_experience.info()
```

```python
# transform data city_development_index

city.sample(10)
city['City'] = city['City'].astype('string')
city.info()
```

```python
# Transform Employment
employment.info()
employment.sample(10)
```

```python
#import vao data warehouse
db_path = 'data_warehouse.db'
engine = create_engine(f'sqlite:///{db_path}')
employment.to_sql('employment', con=engine, if_exists='replace', index=False)
enrollies_data.to_sql('enrollies', con=engine, if_exists='replace', index=False)
city.to_sql('city_development_index', con=engine, if_exists='replace', index=False)
education_data.to_sql('education', con=engine, if_exists='replace', index=False)
work_experience.to_sql('work_experience', con=engine, if_exists='replace', index=False)
training_hours.to_sql('training_hours', con=engine, if_exists='replace', index=False)

```
